```python
q1=prices.quantile(0.25)
q3=prices.quantile(0.75)
```

```python
iqr=q3-q1
iqr
```

this is the inter quartile range,we now find the outliers from the IQR rule

```python
prices_filtered = prices[~((prices < (q1 - 1.5 * iqr)) |(prices > (q3 + 1.5 * iqr)))]
prices_filtered
```

i removed the entries greater than q3+1.5*iqr 

```python
plt.hist(prices_filtered,bins=40)
```

this histogram is much better as it has all houses with prices below 400000 units,this one is much more informative

```python
prices3=prices[prices<1000000]
```
